```python
import pandas as pd
```

```python
def read_data():
    df = pd.read_csv("../data/OK/aggregate_members_oathkeepers.csv")
    return df.drop(columns=["address", "zip", "number_1", "number_2"])
```

```python
df = read_data()
df
```

```python
df.state.unique()
```

```python
def split_names(df):
    names = df.name.str.lower().str.strip().str.extract(r"(\w+-?\w+?) ?(?:(\w+)\.? )?(.+)")

    df.loc[:, "first_name"] = names[0]
    df.loc[:, "middle_name"] = names[1]
    df.loc[:, "last_name"] = names[2]
    return df.drop(columns=["name"])

df = df.pipe(split_names)
```

```python
df.loc[:, "notes"] = df.notes.str.lower().str.strip()
df = df.fillna("")
```
